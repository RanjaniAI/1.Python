```python
#Body Mass Index 
Bmi=int(input("Enter the BMI index:"))
if(Bmi<18):
    print("Under Weight")
elif(Bmi<25):
    print("Healthy weight")
elif(Bmi<30):
    print("Over weight")
elif(Bmi<35):
    print("Obese")
elif(Bmi<40):
    print("Severe Obese")
else:
    print("Morbidly Obese")
    
```
