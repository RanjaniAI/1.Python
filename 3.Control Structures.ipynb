```python
age=int(input("Enter the age:"))
if(age<18):
    print("Childern")
elif(age<35):
    print("Adult")
elif(age<59):
    print("Citizen")
else:
    print("Senior citizen")
```

```python
agelist=[14,25,35,40,59,79]
for age in agelist:
    if(age<18):
        print("Childern")
    elif(age<35):
        print("Adult")
    elif(age<59):
        print("Citizen")
    else:
        print("Senior citizen")
    
```
